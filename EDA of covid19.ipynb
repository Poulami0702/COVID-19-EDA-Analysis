```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
plt.figure(figsize=(25,25))
sns.barplot(x='Population ', y='Country', data=df)
plt.title('Country vs Population',fontsize=25)
plt.xlabel('Population',fontsize=25)
plt.ylabel('Country',fontsize=25)
```

```python
plt.figure(figsize=(25,25))
sns.barplot(x='Total Confirmed Cases/Population', y='Country', data=df)
plt.title('Country vs Total Confirmed Cases/Population',fontsize=25)
plt.xlabel('Total Confirmed Cases/Population',fontsize=25)
plt.ylabel('Country',fontsize=25)
```

```python
plt.figure(figsize=(25,25))
sns.barplot(x='Recovered/Population', y='Country', data=df)
plt.title('Country Vs Recovered/Population',fontsize=25)
plt.xlabel('Recovered/Population',fontsize=25)
plt.ylabel('Country',fontsize=25)
```

```python
plt.figure(figsize=(25,25))
sns.barplot(x='Total Deaths/Population', y='Country', data=df)
plt.title('Country Vs Total Deaths/Population',fontsize=25)
plt.xlabel('Total Deaths/Population',fontsize=25)
plt.ylabel('Country',fontsize=25)
```

```python
sns.distplot(df['Recovered/Population'])
```

```python
sns.distplot(df['Total Deaths/Population'])
```

```python
sns.jointplot(df['Recovered/Population'], df['Total Deaths/Population'], kind="reg")
```

```python
#Population Density (Square KM)
plt.figure(figsize=(25,25))
sns.barplot(x='Population Density (Square KM)', y='Country', data=df)
plt.title('Country Vs Population Density (Square KM)',fontsize=25)
plt.xlabel('Population Density (Square KM)',fontsize=25)
plt.ylabel('Country',fontsize=25)
```

```python
#No. of Doctors per thousands
#Population Density (Square KM)
plt.figure(figsize=(25,10))
sns.barplot(x='No. of Doctors per thousands', y='Population Density (Square KM)', data=df)
plt.title('No. of Doctors per thousands Vs Population Density (Square KM)',fontsize=25)
plt.ylabel('Population Density (Square KM)',fontsize=25)
plt.xlabel('No. of Doctors per thousands',fontsize=25)
```
